Documents

```python
from nltk.corpus import state_union,gutenberg,abc,inaugural

doc1=state_union.raw("1963-Kennedy.txt")
doc2=gutenberg.raw("milton-paradise.txt")
doc3=abc.raw("science.txt")
doc4=inaugural.raw("1789-Washington.txt")
docs=[doc1,doc2,doc3,doc4]
```

**INDEXATION**

-tokenisation:

```python
from nltk.tokenize import word_tokenize
docwords=[]
for doc in docs:
  docwords.append(word_tokenize(doc))
print(docwords[0])

  
```

-Elimination des mots vides:

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
filtered_docs=[]
for words in docwords:
  filtered_doc=[]
  for word in words:
    if word not in stop_words:
      filtered_doc.append(word)
  filtered_docs.append(filtered_doc)
print(filtered_docs[1])
```

-Racinisation:

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_docs=[]
for filtered_doc in filtered_docs:
  stemmed_doc=[]
  for word in filtered_doc:
    stemmed_doc.append(stemmer.stem(word))
  stemmed_docs.append(stemmed_doc)
print(stemmed_docs[2])
```

-Lemmatisation:

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_docs=[]
for doc in filtered_docs:
  lemmatized_doc=[]
  for word in doc:
    lemmatized_doc.append(lemmatizer.lemmatize(word))
  lemmatized_docs.append(lemmatized_doc)
print(lemmatized_docs[3])
```
